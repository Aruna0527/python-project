# **Analysis to be done**


* **Descriptive Analysis (summarizing the dataset)**

    insights into employee performance, engagement, and retention.

* **Diagnostic Analysis (finding reasons behind patterns)**
    
    If an organization's employee turnover rate has increased, diagnostic analytics can help HR teams identify the reasons behind this trend.
    
    By analyzing data from exit interviews, performance evaluations, and other sources, HR teams can gain insights into why employees are leaving and what can be done to improve retention.

* **Predictive Analysis (optional if forecasting required using basic steps)**

    Predictive analytics can be incredibly valuable for HR teams, as it allows them to anticipate potential issues and take proactive measures to address them.

* **Prescriptive Analysis (recommendations for business decisions)**

    Prescriptive analytics can help HR teams make more informed decisions about everything from hiring and onboarding to employee development and retention.

## Stage 2 – Data Cleaning and Pre-processing



```python
numeric_columns = ['no_of_trainings', 'age', 'previous_year_rating', 'length_of_service', 'awards_won', 'avg_training_score']

df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

df = df.dropna(subset=numeric_columns)
print("Number of rows after removing irrelevant data type values:", len(df))
```

```python
df.duplicated()
```

```python
df = df.drop_duplicates()
print("Number of rows after removing duplicates:", len(df))
```

```python
df = df[df['no_of_trainings'] >= 0]
df = df[(df['previous_year_rating'] >= 1) & (df['previous_year_rating'] <= 5)]
df = df[df['length_of_service'] >= 0]
df = df[(df['KPIs_met_more_than_80'] == 0) | (df['KPIs_met_more_than_80'] == 1)]
df = df[df['awards_won'] >= 0]
df = df[(df['avg_training_score'] >= 30) & (df['avg_training_score'] <= 100)]
print("Number of rows after removing irrelevant values:", len(df))
```

```python
df = df.fillna({'education':'Not available','previous_year_rating':0})
print("Number of rows after filling missing values:", len(df))

```

## Insights from pivot table

**By combining previous_year_rating with awards_won or promoted_in_last_year, you can identify if high ratings consistently lead to promotions.**


**Performance Stability: If the data includes multiple years of ratings, a pivot table can show if an employee's performance is improving, declining, or stable over time.**

```python
performance_pivot = pd.pivot_table(
    df,
    values=['previous_year_rating', 'employee_id'], # Values to aggregate
    index=['department'], # Rows of the pivot table
    aggfunc={'previous_year_rating': np.mean, 'employee_id': len}, # Aggregation functions
    fill_value=0 # Replace NaN with 0
)
# Print the resulting pivot table
print("Employee Performance Pivot Table:")
print(performance_pivot.round(2))

# Another example: Average monthly income by department and performance rating
income_pivot = pd.pivot_table(
    df,
    values=['avg_training_score'],
    index=['department'],
    columns=['previous_year_rating'], # Columns of the pivot table
    aggfunc=np.mean,
    fill_value=0
)

print("\nAverage Monthly Income Pivot Table:")
print(income_pivot.round(2))
```

## Correlation Heatmap  


 **The image displays a correlation matrix (heatmap), which is used to visualize the strength and direction of linear relationships between several numerical variables**

which cross-references different regions with various professional functions, indicating performance ratings on a scale from approximately 2.5 to 4.5.

**1. Highest Rating:**

Region\_25 has the highest individual rating of 4.80 for R&D.

**2. Lowest Rating:**

Region\_28 has the lowest individual rating of 2.33 for Procurement.

**3. Highest Average:**

Region\_10 has consistently high ratings across all functions.

**4. Lowest Average:**

Region\_4 has consistently lower ratings compared to the other regions

Heatmap: Top five regions by previous year ratings.




```python
top_5_regions = (
    df.groupby('region')['previous_year_rating']
    .mean()
    .nlargest(5)
    .index
)
df_top5 = df[df['region'].isin(top_5_regions)]
##Prepare data for heatmap: Pivot table (Average ratings by region and year)
heatmap_data = df_top5.pivot_table(
    values='previous_year_rating',
    index='region',
    columns='department',
    aggfunc='mean'
)
plt.figure(figsize=(10, 6))
sns.heatmap(
    heatmap_data,
    annot=True,
    cmap='YlGnBu',
    fmt=".2f",
    linewidths=.5
)
plt.title('Top 5 Regions by Previous Year Rating (Heatmap)')
plt.ylabel('Region')
plt.xlabel('department')
plt.show()

```
